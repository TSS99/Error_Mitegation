# Creating Confusion Matrix

Preparing the four **circuits** needed to calibrate a readout confusion matrix on real quantum hardware:

$$
\lvert 00 \rangle, \quad \lvert 01 \rangle, \quad \lvert 10 \rangle, \quad \lvert 11 \rangle
$$

Each circuit prepares one computational basis state and measures both qubits. These circuits are are meant to be run on an actual QPU. Once the measurement counts from hardware is found, each state's histogram becomes one column of the confusion matrix

$$
M_{ij} = P(\text{observed } i \mid \text{true } j)
$$



```python
from pyquil.quil import Program
from pyquil.gates import H, CNOT, X, MEASURE


shots = 1024
```

## Circuit 1: Prepare $\lvert 00 \rangle$

Both qubits start in $\lvert 0 \rangle$ by default, so no gates are needed — we only declare the classical register and measure.

```python
p00 = Program()

ro = p00.declare('ro', memory_type='BIT', memory_size=2)

p00 += MEASURE(0, ro[0])
p00 += MEASURE(1, ro[1])

p00.wrap_in_numshots_loop(shots)
```

## Circuit 2: Prepare $\lvert 01 \rangle$

The Pauli-X gate flips a qubit from $\lvert 0 \rangle$ to $\lvert 1 \rangle$. To prepare $\lvert 01 \rangle$ (qubit 0 stays $0$, qubit 1 becomes $1$), we apply $X$ to qubit 1 only.

```python
p01 = Program()

ro = p01.declare('ro', memory_type='BIT', memory_size=2)

p01 += X(1)

p01 += MEASURE(0, ro[0])
p01 += MEASURE(1, ro[1])

p01.wrap_in_numshots_loop(shots)
```

## Circuit 3: Prepare $\lvert 10 \rangle$

To prepare $\lvert 10 \rangle$ (qubit 0 becomes $1$, qubit 1 stays $0$), we apply $X$ to qubit 0 only.

```python
p10 = Program()

ro = p10.declare('ro', memory_type='BIT', memory_size=2)

p10 += X(0)

p10 += MEASURE(0, ro[0])
p10 += MEASURE(1, ro[1])

p10.wrap_in_numshots_loop(shots)
```

## Circuit 4: Prepare $\lvert 11 \rangle$

To prepare $\lvert 11 \rangle$, we apply $X$ to both qubits, flipping each from $\lvert 0 \rangle$ to $\lvert 1 \rangle$.

```python
p11 = Program()

ro = p11.declare('ro', memory_type='BIT', memory_size=2)

p11 += X(0)
p11 += X(1)

p11 += MEASURE(0, ro[0])
p11 += MEASURE(1, ro[1])

p11.wrap_in_numshots_loop(shots)
```
